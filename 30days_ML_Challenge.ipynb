```python
import numpy as np 
import pandas as pd

from xgboost import XGBRegressor
from sklearn import preprocessing, model_selection
from sklearn.ensemble import RandomForestRegressor
from sklearn.preprocessing import OrdinalEncoder
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
```

```python
train_data = pd.read_csv("train.csv")
test_data = pd.read_csv("test.csv")
print("Is there a null value in train data?" , pd.isnull("train_data"))
print("Is there a null value in test data?", pd.isnull("test_data"))
```

```python
print("Split-up of train data types:","\n",train_data.dtypes.value_counts())
print("\n")
print("Split-up of test data types:","\n",test_data.dtypes.value_counts())
print("\n")
print("The columns in the train dataset is as follows:")
```

```python
train_data.columns.to_list()
```

```python
train_data["kfold"]=-1
```

```python
train_data.head(10)
```

```python
kf = model_selection.KFold(n_splits=5, random_state=42, shuffle=True)

for fold, (train_indices, val_indices) in enumerate(kf.split(X=train_data)):
    train_data.loc[val_indices,"kfold"] = fold
```

```python
test_data.head()
```

```python
train_data.to_csv("train_folds_data.csv", index=False)
```

```python
df = pd.read_csv("train_folds_data.csv")
df_test =  test_data
```

```python
df.head()
```

```python
df_test.head()
```

```python
selected_cols = [col for col in df.columns if col not in ("id", "target", "kfold")]
object_cols = [col for col in selected_cols if 'cat' in col]
df_test = df_test[selected_cols]
```

```python
final_scores=[]

for fold in range(5):
    xtrain = df[df['kfold']!=fold].reset_index(drop=True)
    xvalid = df[df['kfold']==fold].reset_index(drop=True)
    xtest = df_test.copy()
    
    ytrain = xtrain.target
    yvalid = xvalid.target
    
    xtrain = xtrain[selected_cols]
    xvalid = xvalid[selected_cols]
       
    #training
    ordinal_encoder = OrdinalEncoder()
    xtrain[object_cols] = ordinal_encoder.fit_transform(xtrain[object_cols])
    xvalid[object_cols] = ordinal_encoder.transform(xvalid[object_cols])
    xtest[object_cols] = ordinal_encoder.transform(xtest[object_cols])
    
    #model building
    model = XGBRegressor(random_state=fold, n_estimators=250, n_jobs=4)
    model.fit(xtrain, ytrain)
    preds_valid = model.predict(xvalid)
    test_preds = model.predict(xtest)
    final_scores.append(test_preds)
    print("The fold & RMSE is:",fold, mean_squared_error(yvalid,preds_valid, squared=False))
```

```python
final_preds=np.mean(np.column_stack(final_scores),axis=1)
```

```python
submission_output = pd.read_csv("sample.csv")
submission_output.target = final_preds
```
